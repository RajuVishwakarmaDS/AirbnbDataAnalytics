```python
df = df[~df['name'].isnull()]
df = df[['room_id', 'survey_id', 'host_id', 'room_type', 'country', 'city', 'borough', 'neighborhood', 'reviews', 'overall_satisfaction', 'accommodates', 'bedrooms', 'price', 'name', 'last_modified', 'latitude', 'longitude', 'location']]
df.isnull().sum()
```

```python
# handle duplicates
def newdatetime(x):
    return datetime.strptime(x, '%Y-%m-%d %H:%M:%S.%f').strftime("%d-%m-%Y %H:%M:%S")

df.drop_duplicates(keep='first' ,inplace=False)
df.last_modified = df.last_modified.apply(lambda x: newdatetime(x))
```

```python
import re

def replaceinvalid(x, repl, replacement):
    return re.sub(repl, replacement, x)

# replace / with ''
df.neighborhood = df.neighborhood.apply(lambda x: replaceinvalid(x, '/', ''))
df.name = df.name.apply(lambda x: replaceinvalid(x, '/', ''))

# replace multiple spaces with single space
df.neighborhood = df.neighborhood.apply(lambda x: replaceinvalid(x, '\s+', ' '))
df.name = df.name.apply(lambda x: replaceinvalid(x, '\s+', ' '))
df.head()
```

```python
# handle country null values
# we cannot parse country values from city names since countries can have same city names
# So we can use lat and long to get the contry names

country = []
for x,y in zip(df.latitude.values, df.longitude.values):
    country.append(str(geolocators.geocode("{}, {}".format(x, y))).split(' ')[-1])
    
df
```
